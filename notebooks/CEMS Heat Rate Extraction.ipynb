```python
import numpy as np
import os
import pandas as pd
import sys
import warnings
warnings.simplefilter('ignore')

#pkg_dir = '/home/mrossol/NaTGenPD'
pkg_dir = '..'
sys.path.append(pkg_dir)
import NaTGenPD as npd
import NaTGenPD.cluster as cluster


#data_dir = '/scratch/mrossol/CEMS'
data_dir = '/Users/mrossol/Downloads/CEMS'
```

# Load SMOKE Data

```python
logger = npd.setup_logger('NaTGenPD.clean', log_level='DEBUG')
for year in (2016, 2017):
    dir_path = os.path.join(data_dir, str(year))
    npd.ParseSmoke.performance_vars(dir_path, year, save=True)
```

# Clean SMOKE Data

```python
logger = npd.setup_logger('NaTGenPD.clean', log_level='DEBUG')

unit_attrs_path = os.path.join(data_dir, 'Mappings', 'emission_01-17-2017.csv')
cc_map = os.path.join(pkg_dir, 'bin', 'cems_cc_mapping.csv')

for year in (2016, 2017):
    smoke_path = os.path.join(data_dir, '{y}/SMOKE_{y}.h5'.format(y=year))
    print('Cleaning {}'.format(os.path.basename(smoke_path)))
    out_path = os.path.join(data_dir, 'SMOKE_Clean_{}.h5'.format(year))
    _ = npd.CleanSmoke.clean(smoke_path, unit_attrs_path=unit_attrs_path, cc_map=cc_map,
                             parallel=True, out_file=out_path)
```

## Combine 2016-2017 clean data

```python
logger = npd.setup_logger('NaTGenPD.handler', log_level='DEBUG')
year_files = []
for year in (2016, 2017):
    year_files.append(os.path.join(data_dir, 'SMOKE_Clean_{}.h5'.format(year)))

comb_file = os.path.join(data_dir, 'SMOKE_Clean_2016-2017.h5')
npd.CEMS.combine_years(comb_file, year_files)
```

# Filter

```python
logger = npd.setup_logger('NaTGenPD.filter', log_level='INFO')
clean_file = os.path.join(data_dir, 'SMOKE_Clean_2016-2017.h5')
out_file = os.path.join(data_dir, 'SMOKE_Filtered_2016-2017.h5')
npd.Filter.run(clean_file, out_file, years=2, parallel=True)
```

# 4th-order Polynomial Fits

```python
logger = npd.setup_logger('NaTGenPD.filter', log_level='INFO')

fit_dir = os.path.join(data_dir, 'CEMS_Fits')
if not os.path.exists(fit_dir):
    os.makedirs(fit_dir)

filtered_file = os.path.join(data_dir, 'SMOKE_Filtered_2016-2017.h5')
npd.PolyFit.run(filtered_file, fit_dir, order=4)
```
